## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python

import pandas as pd
```

```python
df = pd.read_csv("../weather-raw.csv")
display(df.head())
```

```python
import math

df_clean = pd.DataFrame(columns=[])

ids = pd.Series([]) 
year = pd.Series([]) 
month = pd.Series([]) 
day = pd.Series([]) 
tmax = pd.Series([]) 
tmin = pd.Series([]) 

flag = 0
cont = 0
for col in df:
    for row in range(len(df[col])):
        if col.startswith("d") and not math.isnan(df[col][row]):
            if flag==0:
                ids[cont] = df["id"][row]
                year[cont] = df["year"][row]
                month[cont] = df["month"][row]
                day[cont] = int(col[1:])
                tmax[cont] = df[col][row]
                cont +=1
            elif flag==1:
                tmin[cont-1] = df[col][row]
            if flag == 1: flag = 0
            else: flag = 1
        
df_clean.insert(0, "id", ids)
df_clean.insert(1, "year", year)
df_clean.insert(2, "month", month)
df_clean.insert(3, "day", day)
df_clean.insert(4, "tmax", tmax)
df_clean.insert(5, "tmin", tmin)

df_clean.sort_values(by=['month', 'day'], inplace=True)
df_clean.reset_index(drop=True, inplace=True)

display(df_clean)

df_clean.to_csv("./weather.csv")

'''df_clean["tmedium"] = df_clean[['tmax', 'tmin']].copy().mean(axis = 1)

df_clean.plot(kind='line',x='month',y='tmax',color='red')
plt.show()


%matplotlib inline


df_clean.groupby(["month"]).mean().plot()'''
```
